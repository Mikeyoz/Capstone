```python
import requests
import matplotlib.pyplot as plt 
import pandas as pd
import json 
import csv 
%matplotlib inline
```

# concat nashville yelp

```python
bars_nash = pd.read_csv("bars_nash.csv")
```

```python
activities_nash = pd.read_csv("activities_nash.csv")
```

```python
activities_nash.head()
```

```python
gyms_nash = pd.read_csv("gyms_nash.csv")
```

```python
pets_nash = pd.read_csv("pets_nash.csv")
```

```python
restaurants_nash = pd.read_csv("restaurants_nash.csv")
```

```python
shopping_nash = pd.read_csv("shopping_nash.csv")
```

```python
nashville_yelp = pd.concat([bars_nash,restaurants_nash, activities_nash, pets_nash, shopping_nash,gyms_nash])
```

```python
nashville_yelp
```

```python
nashville_yelp.categories.value_counts()
```

```python
nashville_yelp['category_one'] = nashville_yelp['categories'].str.split(',').str[0]
```

```python
nashville_yelp['category_two'] = nashville_yelp['categories'].str.split(',').str[1]
```

```python
nashville_yelp['category_three'] = nashville_yelp['categories'].str.split(',').str[2]
```

```python
nashville_yelp
```

```python
nashville_yelp.to_csv("nashville_yelp.csv")
```
