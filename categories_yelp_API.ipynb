```python
import requests
import matplotlib.pyplot as plt 
import pandas as pd
import json 
import csv 
#from IPython.display import Image
%matplotlib inline
```

```python
api_keys = pd.read_csv('api_keys.csv')
#.ipynb_checkpoints
```

```python
api_key = api_keys.loc[api_keys['endpoint'] == 'yelp', 'key'].values[0]
```

```python
end_point_category = 'https://api.yelp.com/v3/businesses/search'
```

```python
header_category_nash = {'Authorization': 'bearer %s' % api_key}
```

# top bars

```python
params_offset = 50
```

```python
def extract_bars_info(bars):
    review_count = bars['review_count']
    categories =  [a['title'] for a in bars['categories']]
    ratings = bars['rating']
    name = bars['name']
    latitude = bars['coordinates']['latitude']
    longitude = bars['coordinates']['longitude']
    try:
        price = bars['price']
    except:
        price = 'no price'
    zip_code = bars['location']['zip_code']
    output_dict_bars = {'review_count': review_count, 
                   'categories': categories, 
                   'ratings': ratings,
                   'name': name, 
                   'latitude': latitude, 
                   'longitude': longitude, 
                   'price': price, 
                   'zip_code': zip_code}
    return output_dict_bars
```

```python
bar_list = []
for increment in range(0, 500, 50) : 
    print(increment)
    params_category_bar = {'location': 'Nashville', 'limit': 50, 'categories': 'bars', 'offset': increment}

    response_business_bar = requests.get(url = end_point_category, params = params_category_bar, headers = header_category_nash)
    print(response_business_bar)
    for bar in response_business_bar.json()['businesses']:
        bar_dict = extract_bars_info(bar)
        bar_list.append(bar_dict)
        params_offset += 50
```

```python
bars_nash = pd.DataFrame(bar_list)
```

```python
bars_nash
```

```python
bars_nash.shape
```

```python
bars_nash.to_csv('bars_nash.csv')
```
